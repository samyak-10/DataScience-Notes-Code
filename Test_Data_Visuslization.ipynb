```python
# Q7. Write a Pandas program to display most frequent value in a given 
# series and replace everything else as 'Other' in the series.
```

```python
series=pd.Series([1,2,2,3,3,3,4])
mf=series.mode()[0]
print("Most Frequent:",mf)
new_series=series.apply(lambda x:x if x==mf else 'Other')
print(new_series)
```

```python
# Q8. Write a Pandas program to get the items of a given series not 
# present in another given series.
```

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([4,5,6,7])
print(s1[~s1.isin(s2)])
```

```python
# Q11. Write a Python program to plot several lines with different 
# format styles in one command using array.
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 10)
y = np.array([1, 3, 3, 21, 3, 4, 3, 4, 3])

plt.plot(x, x, 'r--', label='Red Dashed')
plt.plot(x, y, 'g-.', label='Green Dash-dot')  # Now using x and y
plt.plot(x, x * 3, 'b:', label='Blue Dotted')
plt.legend()
plt.title("Multiple Line Styles")
plt.show()
```

```python
# Q12. Write a Python programming to display a bar chart of the 
# popularity of programming Languages. 
# Sample data: 
# Programming languages: Java, Python, PHP, JavaScript, C#, C++ 
# Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
```

```python
languages=['Java','Python','PHP','JavaScript','C#','C++']
popularity=[22.2,17.6,8.8,8,7.7,6.7]
plt.bar(languages,popularity,color='skyblue')
plt.title("Popularity of Programming Languages")
plt.xlabel("Languages")
plt.ylabel("Popularity")
plt.show()
```

```python
### Q13 Write a Python program to create bar plot from a DataFrame. 
# Sample Data Frame: 
# a b c d e 
# 2 4,8,5,7,6 
# 4 2,3,4,2,6 
# 6 4,7,4,7,8 
# 8 2,6,4,8,6 
# 10 2,4,3,3,2
```

```python
# Q14. Write a Python programming to create a pie chart of the 
# popularity of programming Languages. 
# Sample data: 
# Programming languages: Java, Python, PHP, JavaScript, C#, C++ 
# Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
```

```python
languages=['Java','Python','PHP','JavaScript','C#','C++']
popularity=[22.2,17.6,8.8,8,7.7,6.7]
plt.pie(popularity,labels=languages,autopct='%1.1f%%',startangle=140)
plt.title("Popularity of  Languages (Pie Chart)")
plt.axis('equal')
plt.show()
```
